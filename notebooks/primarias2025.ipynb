```python
import pandas as pd

# --- üöÄ Cargar el archivo de m√©tricas actualizado ---
try:

    df_metricas = pd.read_csv('../results/exp_metricas_polarizacion.csv', sep=';')
    print("‚úÖ Archivo de m√©tricas cargado exitosamente.")
except FileNotFoundError as e:
    print(f"Error al cargar archivo: {e}. Aseg√∫rate de que el script 'analisis_comparativo.py' se haya ejecutado correctamente.")

# --- üéØ AN√ÅLISIS FINAL DE M√âTRICAS DEL PAPER ---

print("\n===============================================")
print(" üìä M√âTRICAS DE POLARIZACI√ìN (Seg√∫n Paper)")
print("===============================================")

# Funci√≥n robusta para imprimir m√©tricas
def print_metric(label, value):
    if isinstance(value, (int, float)) and not pd.isna(value):
        if 'SIZE' in label or 'Nodos' in label or 'K' == label:
             print(f"  - {label}: {int(value)}") # Formato entero para conteos
        else:
             print(f"  - {label}: {value:.4f}") # Formato decimal para ratios/√≠ndices
    else:
        print(f"  - {label}: N/A")

# Itera sobre cada fila (una por dataset)
for index, row in df_metricas.iterrows():
    print(f"\n--- An√°lisis para el Dataset: {row['Dataset']} ---")
    
    # M√©tricas Principales de Polarizaci√≥n
    print_metric("POL (Puntaje de Polarizaci√≥n)", row.get('POL'))
    print_metric("BA-POL (Polarizaci√≥n Balanceada por Tama√±o)", row.get('BA-POL'))
    print_metric("ISO (Aislamiento de la comunidad polarizada)", row.get('ISO'))
    
    # M√©tricas de Estructura de Cl√∫ster
    print_metric("SIZE (Nodos en los cl√∫steres S1 y S2)", row.get('SIZE'))
    print_metric("BAL (Balance entre tama√±os de cl√∫steres)", row.get('BAL'))
    print_metric("K (N√∫mero de cl√∫steres no vac√≠os)", row.get('K'))

    # M√©tricas de Cohesi√≥n y Separaci√≥n (basadas en enlaces +/-)
    print_metric("CC+ (Cohesi√≥n Interna: enlaces + intra-cl√∫ster)", row.get('CC+'))
    print_metric("CC- (Separaci√≥n: enlaces - inter-cl√∫ster)", row.get('CC-'))
    
    # M√©tricas de Coeficiente de Clustering (NUEVAS)
    print_metric("CC_intra_S1 (Clustering Coeff. en S1)", row.get('CC_intra_S1'))
    print_metric("CC_intra_S2 (Clustering Coeff. en S2)", row.get('CC_intra_S2'))

print("\n--- Tabla completa de M√©tricas ---")
display(df_metricas)
```
